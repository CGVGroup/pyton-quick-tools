```python
quartile_rank = {"Q1": 1, "Q2": 2, "Q3": 3, "Q4": 4}
# Function to parse and extract the best quartile values for each area
def extract_best_quartile(row):
    # Split the Areas and Categories based on semicolon (`;`) separators
    areas = [area.strip() for area in row['Areas'].split(';')]
    categories = [category.strip() for category in row['Categories'].split(';')]
    
    # Initialize the output dictionary for the new columns
    row_data = {'Computer Science': None, 'Psychology': None, 'Medicine': None, 'Neuroscience' : None}
    
    # Temporary storage for the best quartiles found
    best_quartiles = {'Computer Science': None, 'Psychology': None, 'Medicine': None, 'Neuroscience' : None}

    # Loop through each area and its corresponding categories
    for area, category_group in zip(areas, categories):
        # Split the categories further by comma if they belong to the same area
        category_list = [category.strip() for category in category_group.split(',')]
        
        # Track the best quartile for the current area
        best_quartile = None
        
        for category in category_list:
            # Use regex to extract the quartile
            match = re.search(r'\((Q[1-4])\)', category)
            if match:
                quartile = match.group(1)
                # If no best quartile found yet or a better (higher) quartile is found
                if best_quartile is None or quartile_rank[quartile] < quartile_rank[best_quartile]:
                    best_quartile = quartile

        # Assign the best quartile to the correct area if it matches the desired areas
        if area == 'Computer Science':
            row_data['Computer Science'] = best_quartile
        elif area == 'Psychology':
            row_data['Psychology'] = best_quartile
        elif area == 'Medicine':
            row_data['Medicine'] = best_quartile
        elif area == 'Neuroscience':
            row_data['Neuroscience'] = best_quartile

    # Return the extracted best quartiles as a Series
    return pd.Series(row_data)

# Apply the function to each row and create the new columns for the best quartiles
scimago_db_df[['Computer Science', 'Psychology', 'Medicine', 'Neuroscience']] = scimago_db_df.apply(extract_best_quartile, axis=1)
```
